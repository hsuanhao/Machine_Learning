# Predict profits for a food truck

I implement $\bf \text{linear regression}$ with one variable to predict profits for a food truck. Assume the CEO of a restaurant franchise is considering different cities for opening a new outlet. The chain already has trucks in various cities, so we study the relation between profits and populations from the cities. We would like to use the data to help select which city to expand to next. In this work, I mainly write codes in $\bf \text{MATLAB/Octave}$ to do data analysis.

## 2. Model and Algorithm

From above figure, the profit seems to have linear relation to the population of city so I assume the variables $y$ (profits) and $x$ (population) so the hypothesis function $h_\theta (x)$ is given by the following linear model:

$$
h_\theta (x) = \theta_0 x_0 + \theta_1 x_1,
$$

where $x_0 = 1$ and $x_1$ denotes population. By using $\bf{\text{batch gradient descent algorithm}}$, we fit the linear regression parameter $\theta$ to the dataset by minimizing the cost function,

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \left(h_\theta(x^{(i)}) - y^{(i)} \right)^2,
$$

where $m$ indicates number to training samples. In batch gradient descent, each iteration performs the update

$$
\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}.
$$

After each step of gradient descent, parameter $\theta_j$ come closer to the optimal values that will achieve the lowest cost $J(\theta)$. 

## 3. Results and Analysis

When one uses gradient descent algorithm to find minimum value of cost function $J(\theta)$, after each iteration, $J(\theta)$ must descrease with each step. Also, one has to choose appropriate learning rate $\alpha$ such that $J(\theta)$ can converge rapidly. In the below figure, it shows that $J(\theta)$ is bowl-shaped and has a global minimum. The second plot shows the contour plot of $J(\theta)$ and denotes where the global minimum found after training. 

```python
from IPython.display import Image
Image(filename='Figure/cost_func_linear.png')
```

After 1500 iterations, $J(\theta)$ converge to 4.4834. When $J(\theta)$ converges to 4.4834, the optimal point $(\theta_0,\theta_1) = (-3.630291, 1.166362)$. Thus, the hypothesis function $h_\theta(x)$ leads to

$$
h_\theta (x) = -3.630291 + 1.166362 x,
$$

where $x$ denotes the population in a city and $h_\theta(x)$ indicates the profits of an outlet. In the following figure, it shows how well the linear model fit in the dataset. 

```python
from IPython.display import Image
Image(filename='Figure/linear_result.png')
```

It is clear that the more population of a city, the more profits of an outlet in that city. After obtaining $\theta$, one can predict the profits of an outlet in a city with given population. For example, we predict a profit of \$4519.77 for population of 35,000. For population of 70,000, we predict a profit of \$45342.45. Thus, opening a new outlet at a city, having more population, would be a best choice according to the analysis of data.
