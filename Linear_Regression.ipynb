# Predict profits for a food truck

We implement $\bf \text{linear regression}$ with one variable to predict profits for a food truck. Assume the CEO of a restaurant franchise is considering different cities for opening a new outlet. The chain already has trucks in various cities, so we study the relation between profits and populations from the cities. We would like to use the data to help select which city to expand to next. # 2. Model and Algorithm

Assume the variables $y$ (profits) and $x$ (population) so the hypothesis function $h_\theta (x)$ is given by the following linear model:
\begin{align}
h_\theta (x) = \theta_0 x_0 + \theta_1 x_1,
\end{align}
where $x_0 = 1$ and $x_1$ denotes population. By using $\bf{\text{batch gradient descent algorithm}}$, we fit the linear regression parameter $\theta$ to the dataset by minimizing the cost function,
\begin{align}
J(\theta) = \frac{1}{2m} \sum_{i=1}^m \left(h_\theta(x^{(i)}) - y^{(i)} \right)^2,
\end{align}
where $m$ indicates number to training samples.
